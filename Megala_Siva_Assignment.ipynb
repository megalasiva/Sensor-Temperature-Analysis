```python
import numpy as np
data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1], [17.3, 22.8, 30.5, 26.4, 21.9], [33.1,29.6, 18.4, 24.0, 27.8], [20.2, 23.5, 31.9, 28.1, 22.6]])
#Task 1: Print Shape and Mean per station
print("Shape of Data:" , data.shape)
mean_per_station = data.mean(axis=1)
print("Mean temperature per station:" , mean_per_station)
mask = data > 28.0
temps_above_28 = data[mask]
print("Temperatures readings above 28.0°C (1D Array):" , temps_above_28)
normalized = (data - data.min()) / (data.max() - data.min())
print("Normalized data (rounded to 2 decimal places):")
print(np.round(normalized, 2))


```
